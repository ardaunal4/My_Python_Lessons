```python
#Example of split() function
name_list = ["Hasan Dede", "aa", "Kadir KAN", "Baris DAN", "Arda BAN", "Suleyman CAN", "Aysegul DAN", "Hasan Burkan"]
data = pd.DataFrame({"Name_List":name_list})
seperate = data.Name_List[data.Name_List != 'aa'].str.split()
seperate
```

```python
a, b = zip(*seperate)
a
b
```

```python
name_list2 = a + b
name_list2 # They become a tuple
```

```python
name_count = Counter(name_list2) # As you can see the Counter function counts names of values
name_count
```

```python
# Most common 15 Name or Surname of killed people
separate = kill.name[kill.name != 'TK TK'].str.split() 
separate
```

```python
a,b = zip(*separate)                    
name_list = a+b                         
name_list
```

```python
name_count = Counter(name_list)          
name_count
```

```python
most_common_names = name_count.most_common(15) 
most_common_names
```

```python
x,y = zip(*most_common_names) # split tuples in the list x -> name , y -> number of name
x,y = list(x),list(y)
x
```

```python
plt.figure(figsize = (15,10))
ax= sns.barplot(x = x, y = y, palette = sns.cubehelix_palette(len(x))) # platte uzunluga gore degisen renk atiyor
plt.xlabel('Name or Surname of killed people')
plt.ylabel('Frequency')
plt.title('Most common 15 Name or Surname of killed people')
```
