```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Create 3 different classes for unsupervised learning algorithm
# Class1
x1 = np.random.normal(25, 5, 1000)
y1 = np.random.normal(25, 5, 1000)

# Class2
x2 = np.random.normal(55, 5, 1000)
y2 = np.random.normal(60, 5, 1000)

# Class3
x3 = np.random.normal(55, 5, 1000)
y3 = np.random.normal(15, 5, 1000)
```

```python
x = np.concatenate((x1, x2, x3), axis = 0)
y = np.concatenate((y1, y2, y3), axis = 0)
```

```python
dictionary = {"x":x, "y":y}
data = pd.DataFrame(dictionary)
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)
plt.title("Created Data")
plt.show()
```

```python
wcss = [] # (wcss : The sum of squared distance between) Quality of clustering according cluster number k list

for k in range(1,15):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
```

```python
# To see which is the best k value
plt.figure(figsize = (10, 8))
plt.plot(range(1, 15), wcss, "*-") 
plt.xlabel("number of k (cluster) value")
plt.ylabel("wcss")
plt.show() 
```

```python
kmeans2 = KMeans(n_clusters = 3) # Breaking point is 3 which is the best for k-means clustering
clusters = kmeans2.fit_predict(data)
data["label"] = clusters
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(data.x[data.label == 0 ], data.y[data.label == 0], color = "red")
plt.scatter(data.x[data.label == 1 ], data.y[data.label == 1], color = "green")
plt.scatter(data.x[data.label == 2 ], data.y[data.label == 2], color = "blue")
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], color = "black") # Show the center of clusters
plt.show()
```
